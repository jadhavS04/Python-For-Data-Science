using matplotlib with numpy

```python
x=np.arange(1,11)
y=np.arange(10,110,10)

plt.figure(figsize=(6,6))
plt.plot(x,y,'r--')
plt.show()
```

```python
x_sin=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x_sin)

print(y_sin)
```

```python
x_sin=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x_sin)

print(y_sin)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_sin,y_sin)
plt.title('sin curve')
plt.show()
```

```python
x_cos=np.arange(0,2*np.pi,0.1)
y_cos =np.cos(x_cos)

plt.figure(figsize=(6,6))
plt.plot(x_cos,y_cos)
plt.title('cos curve')
plt.show()
```

```python
x_tan=np.arange(0,2*np.pi,0.1)
y_tan =np.tan(x_tan)

plt.figure(figsize=(6,6))
plt.plot(x_tan,y_tan)
plt.title('tan curve')
plt.show()
```

```python
x_cot=np.arange(0,2*np.pi,0.1)
y_cot =1/np.tan(x_cot)

plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot)
plt.title('cot curve')
plt.show()
```

subplot

```python
plt.figure(figsize=(6,6))
plt.subplot(2,2,1)
plt.plot(x_sin,y_sin,'r-')
plt.title('sin curve')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,'b-')
plt.title('cos curve')

plt.subplot(2,2,3)
plt.plot(x_tan,y_tan,'g-')
plt.title('tan curve')

plt.subplot(2,2,4)
plt.plot(x_cot,y_cot,'w-')
plt.title('cot curve')

plt.show()
```

```python
import numpy as np

# Step 1: Represent the data in a 10x2 array
data = np.array([
    [1, 6012],
    [2, 7079],
    [3, 6886],
    [4, 7230],
    [5, 4598],
    [6, 5564],
    [7, 6971],
    [8, 7763],
    [9, 8032],
    [10, 9569]
])

# Step 2: Add 2000 steps to each observation
data[:, 1] += 2000

# Step 3: Return the steps walked if more than 9000
steps_more_than_9000 = data[data[:, 1] > 9000]

# Step 4: Print an array containing steps walked in sorted order
sorted_steps = np.sort(data[:, 1])

# Output the results
print("Data with added steps:\n", data)
print("\nSteps walked more than 9000:\n", steps_more_than_9000)
print("\nSorted steps walked:\n", sorted_steps)
```
