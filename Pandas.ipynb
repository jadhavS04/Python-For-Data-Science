```python
import pandas as pd
from google.colab import files

# Upload the file
uploaded = files.upload()


# Load the CSV file
df = pd.read_csv('rainfall1627650742214 (1).csv')
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows with missing values
df = df.dropna()

# Verify that there are no missing values left
print(df.isnull().sum())
```

```python
# Find the district with the highest annual rainfall
max_rainfall_district = df.loc[df['ANNUAL'].idxmax()]
print("District with the highest annual rainfall:")
print(max_rainfall_district[['DISTRICT', 'ANNUAL']])
```

```python
# Group by state and sum annual rainfall
state_rainfall = df.groupby('STATE_UT_NAME')['ANNUAL'].sum()

# Sort and display the top 5 states
top_5_states = state_rainfall.sort_values(ascending=False).head(5)
print("Top 5 states with the highest annual rainfall:")
print(top_5_states)
```

```python
# Drop the specified columns
columns_to_drop = ['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']
df = df.drop(columns=columns_to_drop)

# Verify the columns are dropped
print(df.columns)
```

```python
# Create a pivot table for state-wise mean rainfall
statewise_mean_rainfall = df.pivot_table(index='STATE_UT_NAME', values=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'], aggfunc='mean')

print("State-wise mean rainfall for all months:")
print(statewise_mean_rainfall)
```

```python
# Count the number of districts per state
district_count = df['STATE_UT_NAME'].value_counts()

print("Count of districts in each state:")
print(district_count)
```

```python
# Group by state and find the district with the highest May rainfall
highest_may_rainfall = df.loc[df.groupby('STATE_UT_NAME')['MAY'].idxmax(), ['STATE_UT_NAME', 'DISTRICT', 'MAY']]

print("District with the highest rainfall in May for each state:")
print(highest_may_rainfall)
```
